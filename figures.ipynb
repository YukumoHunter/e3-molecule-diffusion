```python
import matplotlib.pyplot as plt

# Define the values and variances
values = [10, 15]
variances = [2, 3]

# Define labels for bars
labels = ['Value 1', 'Value 2']

# Define positions for bars
x_pos = range(len(values))

# Create the bar chart with error bars
plt.bar(x_pos, values, align='center', alpha=0.5)
for i in range(len(values)):
    plt.errorbar(x_pos[i], values[i], yerr=variances[i], fmt='k_', capsize=10)
    plt.text(x_pos[i], values[i] + max(variances) + 0.5, f'{values[i]} ± {variances[i]}', ha='center')

# Set y-axis limits
plt.ylim(0, max(values) + max(variances) + 2)

# Add labels, title, and legend
plt.xticks(x_pos, labels)
plt.ylabel('Values')
plt.title('Bar Chart with Variances')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the values and variances for number1
values1 = [10, 15]
variances1 = [2, 3]

# Define the values and variances for number2
values2 = [12, 18]
variances2 = [3, 4]

# Define labels for bars
labels = ['Pytorch', 'JAX']

# Define positions for bars
x_pos = range(len(values1))

# Create a 1x2 subplot
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot number1 in the first subplot
axs[0].bar(x_pos, values1, align='center', alpha=0.5)
for i in range(len(values1)):
    axs[0].errorbar(x_pos[i], values1[i], yerr=variances1[i], fmt='k_', capsize=10)
    axs[0].text(x_pos[i], values1[i] + max(variances1) + 0.5, f'{values1[i]} ± {variances1[i]}', ha='center')
axs[0].set_ylim(0, max(values1) + max(variances1) + 2)
axs[0].set_xticks(x_pos)
axs[0].set_xticklabels(labels)
axs[0].set_ylabel('nLL')
axs[0].set_title('nLL per model')

# Plot number2 in the second subplot
axs[1].bar(x_pos, values2, align='center', alpha=0.8, color = "salmon")
for i in range(len(values2)):
    axs[1].errorbar(x_pos[i], values2[i], yerr=variances2[i], fmt='k_', capsize=10)
    axs[1].text(x_pos[i], values2[i] + max(variances2) + 0.5, f'{values2[i]} ± {variances2[i]}', ha='center')
axs[1].set_ylim(0, max(values2) + max(variances2) + 2)
axs[1].set_xticks(x_pos)
axs[1].set_xticklabels(labels)
axs[1].set_ylabel('Time (s)')
axs[1].set_title('Time per model')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coordinates of the points
x_values = [1, 2, 3]
y_values = [2, 3, 1]

# Fit a polynomial regression curve (degree 2) to the data
coefficients = np.polyfit(x_values, y_values, 2)
poly_function = np.poly1d(coefficients)
x_fit = np.linspace(min(x_values), max(x_values), 100)
y_fit = poly_function(x_fit)

# Create a scatter plot with SeaGreen color for points
plt.scatter(x_values, y_values, color='SeaGreen', label='Points')

# Add the polynomial regression curve
plt.plot(x_fit, y_fit, color='SeaGreen', alpha=0.6, label='Polynomial regression (degree 2)')

# Add grid with alpha = 0.5 in light gray
plt.grid(True, color='lightgray', alpha=0.5)

# Set integer locator for major ticks on x-axis
plt.xticks(range(1, max(x_values) + 1))

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Three Points and Polynomial Regression Curve')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coordinates of the points for Set1
x_values_set1 = [1, 5, 10]
y_values_set1 = [6, 3, 1]

# Fit a polynomial regression curve (degree 2) to Set1
coefficients_set1 = np.polyfit(x_values_set1, y_values_set1, 2)
poly_function_set1 = np.poly1d(coefficients_set1)
x_fit_set1 = np.linspace(min(x_values_set1), max(x_values_set1), 100)
y_fit_set1 = poly_function_set1(x_fit_set1)

# Define the coordinates of the points for Set2
x_values_set2 = [1, 5, 10]  # Example values for demonstration
y_values_set2 = [3, 4, 8]        # Example values for demonstration

# Fit a polynomial regression curve (degree 2) to Set2
coefficients_set2 = np.polyfit(x_values_set2, y_values_set2, 2)
poly_function_set2 = np.poly1d(coefficients_set2)
x_fit_set2 = np.linspace(min(x_values_set2), max(x_values_set2), 100)
y_fit_set2 = poly_function_set2(x_fit_set2)

# Create a scatter plot with SeaGreen color for points in Set1
plt.scatter(x_values_set1, y_values_set1, color='SeaGreen', label='Compilation Time')

# Add the polynomial regression curve for Set1
plt.plot(x_fit_set1, y_fit_set1, color='SeaGreen', alpha=0.6) #, label='Set1 Polynomial regression (degree 2)')

# Create a scatter plot with OrangeRed color for points in Set2
plt.scatter(x_values_set2, y_values_set2, color='Crimson', label='Run time')

# Add the polynomial regression curve for Set2
plt.plot(x_fit_set2, y_fit_set2, color='Crimson', alpha=0.6) #, label='Set2 Polynomial regression (degree 2)')

# Add grid with alpha = 0.5 in light gray
plt.grid(True, color='lightgray', alpha=0.5)

# Set integer locator for major ticks on x-axis
plt.xticks(range(1, max(x_values_set1 + x_values_set2) + 1))

# Add labels and title
plt.xlabel('Number of time steps we jit')
plt.ylabel('Time (s)')
plt.title('Compilation and Run time in terms of the time steps that we jit')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
